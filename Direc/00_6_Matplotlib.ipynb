## Matplotlib

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#line
plt.plot([1, 3, 6, 7, 11])
plt.show()
```

```python
x = np.arange(-2, 2, 0.1) #-2부터 2-간격까지, 0.1간격으로 숫자 생성
x
```

```python
fx = np.exp(x)
plt.plot(x, fx)
plt.show()
```

```python
fx = np.exp(x)
plt.plot(x, fx, ls=':', c='y', lw=3, marker='^', mfc='w')
plt.plot(x, fx + 1, ls='--', marker='x')
plt.plot(x, fx + 2, ls='-', marker='o', ms=5, mec='k', mew=0.5)
plt.title('fx & fx+1')
plt.xlabel('x')
plt.ylabel('fx')
plt.grid()
plt.show()
```

```python
X = np.random.randn(10, 2)
X
```

```python
plt.scatter(X[:, 0], X[:, 1], c='k', marker='o', s=100, lw=10) #c=색, marker=모양
plt.show()
```

```python
data = np.random.normal(0, 1, 5) #정규분포 ( 평균, 표준편차, 개수)
data
```

```python
plt.hist(data, bins=5) #분할 갯수 : 5
plt.show()
```

```python
data = np.random.normal(0, 1, 100000) #10000개
plt.hist(data, bins=100)
plt.show()
```

#### subplot 

```python
x1 = np.linspace(0, 4*3.14)
x2 = np.linspace(0, 2*3.14)
y1 = np.cos(x1)
y2 = np.cos(x2)*(1/2)
```

```python
ax1 = plt.subplot(211)
plt.title('cos & 1/2*cos')
plt.ylabel('y')
plt.plot(x1, y1, 'yo--') #yellow, o, --

ax2 = plt.subplot(212)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x2, y2, 'bo--') #yellow, o, --

plt.tight_layout()
plt.show()
```
