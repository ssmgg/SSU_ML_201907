## Matplotlib

```python
import pandas as pd
import numpy as np

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1, 3, 6, 7, 11])
plt.show()
```

```python
x = np.arange(-2, 2, 0.1)
x
```

```python
fx = np.exp(x)
plt.plot(x, fx)
plt.show()
```

```python
fx = np.exp(x)
plt.plot(x, fx)
plt.plot(x, fx + 1)
plt.show()
```

```python
X = np.random.randn(10, 2)
X
```

```python
plt.scatter(X[:, 0], X[:, 1], c='k', marker='o')
plt.show()
```

```python
data = np.random.normal(0, 1, 10)
data
```

```python
plt.hist(data, bins=5)
plt.show()
```

```python
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=10)
plt.show()
```
